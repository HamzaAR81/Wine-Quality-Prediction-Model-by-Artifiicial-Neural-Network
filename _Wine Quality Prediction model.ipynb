```python
for i in df.columns:                       # making an algorithm to remove outliers from the features:
    x=df[i].describe()
    lst=[]
    for j in df[i]:
        if j > x[1]+(x[2]*3):
            lst.append(x[1]+x[2]*3)
        elif j < x[1]+(x[2]*-3):
            lst.append(x[1]+x[2]*-3)
        else:
            lst.append(j)
    df[i]=lst
        
        
```

# Training the data:

```python
from sklearn.ensemble import RandomForestRegressor # importing the model from library:
rf = RandomForestRegressor()
```

```python
x=df.drop('quality',axis=1) # x is the training set and y is the targeted values
y=df.quality
rf.fit(x,y)
```

```python
df['quality']=df['quality'].astype('int32') # converting the float valuen of qaulity feature in integer:
```

```python
from imblearn.over_sampling import SMOTE # importing the SMOTE method
```

```python
y=df['quality']
smote=SMOTE(sampling_strategy='auto')  # balancing the data by using smote technique
x_sm,y_sm=smote.fit_resample(x,y)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()                          # standardizing the value by using standard scalar 
x_sm[x_sm.columns]=scaler.fit_transform(x_sm[x_sm.columns])
```

```python
x_sm
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_sm,y_sm,test_size=0.3,random_state=42) # spliting the training and test sets:
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y_train=le.fit_transform(y_train) # labeling the values using label encoder
y_test=le.fit_transform(y_test)
```

# Making Neural Network:

```python
# imporing the libraries for making ANN:
import tensorflow 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input,Dense
from tensorflow.keras.layers import Dropout
```

```python
# making neural layers with 3 hidden layers
model = Sequential()
model.add(tensorflow.keras.layers.Input(shape=(11,)))
model.add(tensorflow.keras.layers.Dense(32,activation='relu'))
model.add(tensorflow.keras.layers.Dense(64,activation='relu'))
model.add(tensorflow.keras.layers.Dropout(0.3))
model.add(tensorflow.keras.layers.Dense(128,activation='relu'))
model.add(tensorflow.keras.layers.Dropout(0.3))          
model.add(tensorflow.keras.layers.Dense(6,activation='softmax'))
```

```python
y_train=tensorflow.keras.utils.to_categorical(y_train,6)
y_test=tensorflow.keras.utils.to_categorical(y_test,6)
```

```python
# training the data
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics='accuracy')
```

```python
model.fit(x_train,y_train,epochs=800,verbose=1)
```

#  Prediction of the data:

```python
y_pred=model.predict(x_test) # predicting the values 
```

```python
y_pred=np.argmax(y_pred,axis=1)
y_pred
```

```python
y_test=np.argmax(y_test,axis=1)
```

# Acurracy check:

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
